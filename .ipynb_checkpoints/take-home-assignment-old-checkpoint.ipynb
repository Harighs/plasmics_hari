# Task 1: Coding task - Game of Life


Using only numpy, write a piece of code that achieves the following in the most efficient/fast way:

Come up with an implementation of "Game of Life" (https://en.wikipedia.org/wiki/Conway%27s_Game_of_Life). Given is an input array (2-dimensional binary image) that should be taken as a starting point. From there on, run 42 iterations (i.e. apply the rules that often), return the result in the given arrray and also visualize the outcome.

The rules that should be considered are:

* Any live cell with two or three live neighbours survives.
* Any dead cell with three live neighbours becomes a live cell.
* All other living cells die in the next generation. Similarly, all other dead cells stay dead.

```python
n = np.random.randint(0,2,size=(5, 5)).astype(np.uint8)
n
```

```python
# to get the neighbors in all directions (including the diagonals)
# convolve with an inverse identity kernel - the diagonals
# with a constant padding of 0 to preserve input shape

# kernel_diag would also include the diagonal neighbors

kernel = np.array([[0, 1, 0],
                  [1, 0, 1],
                  [0, 1, 0]]).astype(np.uint8)

kernel_diag = np.array([[1, 1, 1],
                  [1, 0, 1],
                  [1, 1, 1]]).astype(np.uint8)

n_neighbors = cv2.filter2D(n, -1, kernel, borderType=cv2.BORDER_CONSTANT)
n_neighbors
```

```python
# 1. survive requirement: cell alive in previos gen and 2 or 3 live neighbors
# we get those by finding cells that have 2 or 3 live neighbors and multiplying it by the previous
# generation

# 2. becoming a live cell: cell dead but 3 live neighbors

# 3. everything else dead

# --> an OR operation uniting 1 and 2 are who survives, everything else dies
live_with_req_neighbors = (np.where(np.logical_or(n_neighbors == 2, n_neighbors == 3), 1, 0) * n).astype(np.uint8)
dead_with_three_neighbors = (np.where(n_neighbors == 3, 1, 0) * (np.invert(n)) / 255).astype(np.uint8)

survived = cv2.bitwise_or(live_with_req_neighbors, dead_with_three_neighbors)

outcome = np.where(survived, 1, 0)
```

```python
fig, ax = plt.subplots(ncols=2, figsize=(10, 10))
ax = ax.ravel()
ax[0].imshow(n)
ax[1].imshow(outcome)
```

```python
def gol(start):
    
    kernel = np.array([[0, 1, 0],
                  [1, 0, 1],
                  [0, 1, 0]]).astype(np.uint8)
    
    n_neighbors = cv2.filter2D(start, -1, kernel, borderType=cv2.BORDER_CONSTANT)
    
    live_with_req_neighbors = (np.where(np.logical_or(n_neighbors == 2, n_neighbors == 3), 1, 0) * start).astype(np.uint8)
    dead_with_three_neighbors = (np.where(n_neighbors == 3, 1, 0) * (np.invert(start)) / 255).astype(np.uint8)

    survived = cv2.bitwise_or(live_with_req_neighbors, dead_with_three_neighbors)

    outcome = np.where(survived, 1, 0)
    
    return outcome

def gol1(start):
    """
    if n_neighbors == 3, the cell survives anyway;
    so it makes sense to just use that as one condition, avoid unnecessary matrix multiplications,
    and just check for live cells that have 2 live neighbors
    """
    
    kernel = np.array([[0, 1, 0],
                  [1, 0, 1],
                  [0, 1, 0]]).astype(np.uint8)
    
    n_neighbors = cv2.filter2D(start, -1, kernel, borderType=cv2.BORDER_CONSTANT)
    
    # two_neighbors_live = (np.where(n_neighbors == 2, 1, 0) * start).astype(np.uint8)
    # avoiding this matrix multiplication above cuts down the runtime by 50%
    two_neighbors_live = np.where(np.logical_and(n_neighbors == 2, start == 1), 1, 0).astype(np.uint8)
    three_neighbors = np.where(n_neighbors == 3, 1, 0).astype(np.uint8)

    survived = cv2.bitwise_or(two_neighbors_live, three_neighbors)

    outcome = np.where(survived, 1, 0)
    
    return outcome
```
